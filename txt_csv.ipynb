```python
import re

# Replace 'text_file_path' with the actual path to your text file
text_file_path = '/home/nic/Desktop/desk/Jupyter files/WhatsApp Chat with PROFITABLE BUSINESS IDEAS.txt'

with open(text_file_path, 'r', encoding='utf-8') as f:
    text_data = f.read()

def convert_whatsapp_chat_to_csv(text_data):
    # Regex pattern to match message headers
    header_pattern = r"(\d{2}/\d{2}/\d{4}, \d{2}:\d{2} [ap]m) - (\+?\d+): \*(.+)\*"

    # Extract message information
    messages = []
    for match in re.finditer(header_pattern, text_data):
        timestamp = match.group(1)
        phone_number = match.group(2)
        title = match.group(3)

        # Remove phone number from title
        title = title.replace('*', '')

        # Extract content
        content = re.sub(r"\+?\d+$", "", text_data[match.end():])

        # Create message dictionary
        messages.append({
            'timestamp': timestamp,
            'title': title,
            'content': content
        })

    return messages

# Example usage
text_data = """16/01/2022, 8:23 pm - +254 742 304047: *BEST SIDE HUSTLES FOR TEACHERS IN KENYA*

This is the content of the first message. It has all sorts of punctuation marks.

+254701711058

16/01/2022, 8:24 pm - +254 742 304047: *ANOTHER MESSAGE TITLE*

This is the content of the second message. It also has all sorts of punctuation marks.

+254701711058
"""

messages = convert_whatsapp_chat_to_csv(text_data)

# Write messages to CSV file
import csv

with open('whatsapp_chat.csv', 'w', newline='') as csvfile:
    fieldnames = ['timestamp', 'title', 'content']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for message in messages:
        writer.writerow(message)
```

```python
import re

def extract_messages(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.read()

    # Corrected regular expression pattern
    pattern = re.compile(r'\((\d{2}/\d{2}/\d{4}, \d{2}:\d{2}\s*[apAP]\.m)\s*-\s*(.*?)\s*:\s*(.*?)\)', re.DOTALL)

    # Find all matches in the data
    matches = pattern.findall(data)

    # Extract timestamp, title, and content from each match
    messages = []
    for match in matches:
        timestamp, title, content = match
        messages.append({
            'timestamp': timestamp.strip(),
            'title': title.strip(),
            'content': content.strip()
        })

    return messages

# Example usage:
file_path = '/home/nic/Desktop/desk/Jupyter files/WhatsApp Chat with PROFITABLE BUSINESS IDEAS.txt'
messages = extract_messages(file_path)

# Print the extracted information for each message
for message in messages:
    print(f"Timestamp: {message['timestamp']}")
    print(f"Title: {message['title']}")
    print(f"Content: {message['content']}")
    print("="*30)
```

using re.search() to find matches in the entire input text. This function only returns the first match it finds. To find all matches, you should use re.finditer() instead. Additionally, you can iterate over the matches to process each header individually. 

```python
import re

input_text = """
17/01/2022, 12:59 am - +254 742 304047: *GOOD MORNING MEMBERS*
16/01/2022, 8:23 pm - +254 742 304047: *BEST SIDE HUSTLES FOR TEACHERS IN KENYA*
17/01/2022, 12:59 am - +254 742 304047: *GOOD MORNING MEMBERS*
16/01/2022, 8:23 pm - +254 742 304047: *BEST SIDE HUSTLES FOR TEACHERS IN KENYA*
"""

# Remove any unnecessary spaces and line breaks
input_text = input_text.strip()

# Create a regular expression pattern to match the date
date_pattern = r'\b(\d{2}/\d{2}/\d{4})'

# Create a regular expression pattern to match the time and am/pm, including non-breaking space
time_pattern = r'(\d{1,2}:\d{2})\s*(am|pm)'

# Create a regular expression pattern to match the title
title_pattern = r'\*(\w.*\w)\*'

# Use re.finditer() to find all matches in the input text
matches = re.finditer(f'{date_pattern}.*?{time_pattern}.*?{title_pattern}', input_text)

# Process each match individually
for match in matches:
    date_match = re.search(date_pattern, match.group())
    date = date_match.group(1) if date_match else None

    time_match = re.search(time_pattern, match.group())
    time, am_pm = time_match.groups() if time_match else (None, None)

    title_match = re.search(title_pattern, match.group())
    title = title_match.group(1) if title_match else None

    # Combine the date and time manually
    timestamp = f"{date}, {time} {am_pm}" if date and time and am_pm else None

    print(f"Date: {date}")
    print(f"Time: {time} {am_pm}")
    print(f"Title: {title}")
    print("="*30)
```
